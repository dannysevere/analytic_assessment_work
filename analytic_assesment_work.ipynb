```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/dannysevere/Documents/Python/octopi.csv')
```

```python
df.head()
```

```python
df.drop(columns=df.columns[[-1, -2]], inplace=True)
df.head()
```

```python
df[['$', 'Media Spend']] = df['Media Spend'].str.split(pat = '$', expand=True)
df.head()
```

```python
df[['$$', 'Revenue']] = df['Revenue'].str.split(pat = '$', expand=True)
df.head()
```

```python
df.drop(columns=df.columns[[-1, -2]], inplace=True)
df.head()
```

```python
df['Impressions'] = df['Impressions'].replace(',', '', regex=True).astype(float)
```

```python
df[['Media Spend', 'Orders', 'Clicks', 'Revenue', 'New Customers', 'Existing Customers']] = df[['Media Spend', 'Orders', 'Clicks', 'Revenue', 'New Customers', 'Existing Customers']].replace(',', '', regex=True).astype(float)
```

```python
df['CPO'] = df['Media Spend']/df['Orders']
df['AVO'] = df['Revenue']/df['Orders']
df['ROAS'] = df['Revenue'] / df['Media Spend']
df.head()
```

```python
df2 = df[['Revenue', 'Year']].groupby('Year').agg('sum')
print(df2)
```

```python
df2['revenue_growth_rate'] = (df2['Revenue'].pct_change() * 100).fillna(0)

print(df2)
```

```python
df3 = df[['Orders', 'Revenue', 'Year', 'Media Channel']].groupby(['Year', 'Media Channel']).agg('sum')
print(df3)
```

```python
df3['AVO'] = df3['Revenue'] / df3['Orders']
```

```python
print(df3)
```

```python
df4 = df[['Revenue', 'Media Spend', 'Subchannel', 'Media Channel']].groupby(['Media Channel', 'Subchannel']).agg('sum')
print(df4)
```

```python
df4['ROAS'] = df4['Revenue'] / df4['Media Spend']
print(df4)
```

```python
df5 = df[['Clicks', 'Media Spend', 'Subchannel', 'Media Channel']].groupby(['Media Channel', 'Subchannel']).agg('sum')
```

```python
df5['CPC'] = df5['Media Spend']/df5['Clicks']
print(df5)
```
